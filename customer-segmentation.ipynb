```r
# Utilizaremos el algoritmo K-means el cual es utilizado para encontrar subgrupod dentro de una base de datos. El objetivo es agrupar a los clientes con comportamientos similares se agrupen.
## Lo priemro es encontrar el número optimo de cluseters (grupos) para eso existen diferentes metodos : Elbow method , silhouette method, Gap Statistic
```

```r
#Elbow method
library(purrr)
set.seed(123)
# function to calculate total intra-cluster sum of square 
iss <- function(k) {
  kmeans(customer_data[,3:5],k,iter.max=100,nstart=100,algorithm="Lloyd" )$tot.withinss
}

k.values <- 1:10


iss_values <- map_dbl(k.values, iss)

plot(k.values, iss_values,
    type="b", pch = 19, frame = FALSE, 
    xlab="Number of clusters K",
    ylab="Total intra-clusters sum of squares") # De la gráfica lo que nos intereza es ver donde parece el dobles de un codo por lo cual concluimos que por este metodo es 4 cluesters
```

```r
# Tomamos k=6 como el optimo
k6<-kmeans(customer_data[,3:5],6,iter.max=100,nstart=50,algorithm="Lloyd")
k6
```

```r
# Visualizaremos los resultados del cluster usando los dos primeros componentes principales
pcclust=prcomp(customer_data[,3:5],scale=FALSE) #principal component analysis
summary(pcclust)

pcclust$rotation[,1:2]
```

```r
set.seed(1)
ggplot(customer_data, aes(x =Annual.Income..k.., y = Spending.Score..1.100.)) + 
  geom_point(stat = "identity", aes(color = as.factor(k6$cluster))) +
  scale_color_discrete(name=" ",
              breaks=c("1", "2", "3", "4", "5","6"),
              labels=c("Cluster 1", "Cluster 2", "Cluster 3", "Cluster 4", "Cluster 5","Cluster 6")) +
  ggtitle("Segments of Mall Customers", subtitle = "Using K-means Clustering")
```
